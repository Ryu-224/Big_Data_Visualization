## 열 삭제하기

```python
import gdown

gdown.download('https://bit.ly/3RhoNho', 'ns_202104.csv', quiet = False)
```

```python
import pandas as pd

ns_df = pd.read_csv('ns_202104.csv', low_memory = False)
ns_df.head()
```

마지막 'Unnamed:13' 열은 CSV 파일 각 라인의 끝에 콤마(,)가 있어서 판다스가 자동으로 추가한 것!

불필요한 열이므로 삭제하는 것이 좋다!!

```python
ns_book = ns_df.loc[:, '번호' : '등록일자'] # '번호' 열부터 '등록일자' 열까지 전체 행을 선택!
ns_book.head()
```

이미 배운 loc 메소드에 슬라이싱을 사용하면 다음과 같이 '번호' 열부터 '등록일자' 열까지 선택하여 새로운 데이터 프레임을 만들 수 있다.

이 방법이 편리하긴 하지만, 중간에 있는 '부가기호' 열을 제외하고 선택하려면 어떻게 해야할까요?

중간에 있는 열을 제외하려면 슬라이싱을 사용하기 어렵다.

이럴 때는 **불리언 배열**(boolean array)을 사용하면 쉽게 원하는 열만 선택할 수 있다.

```python
print(ns_df.columns)
```

columns 속성은 판다스의 Index 클래스 객체입니다.

이 객체의 원소는 파이썬의 리스트처럼 숫자 인덱스로 참조할 수 있다.

예를들어 첫번째 열 이름인 '번호'를 선택하려면 다음과 같이 쓸 수 있다.

```python
print(ns_df.columns[0])
```

```python
ns_df.columns != 'Unnamed: 13'
```

원소별 비교를 활용하여 'Unnamed: 13' 열이 아닌 것을 표시하는 배열을 만들면 위와 같다!

결과는 넘파이 배열(Numpy array)이다. 지금은 넘파이 배열을 파이썬 리스트와 비슷한 것으로 생각해도 좋다. 원소별 비교 연산으로 불리언 배열을 얻었다.

이를 selected_columns 변수에 저장하고 판다스 데이터프레임의 loc 메소드에 전달하면 True인 열의 행만 선택할 수 있다!

```python
selected_columns = ns_df.columns != 'Unnamed: 13'
ns_book = ns_df.loc[:, selected_columns] # True인 열의 모든 행 선택!
ns_book.head()
```

이번에는 데이터프레임 중간에 있는 '부가기호' 열 제거

```python
selected_columns = ns_df.columns != '부가기호'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

이번에는 데이터프레임의 메소드를 사용하여 열을 삭제!

판다스에서는 데이터프레임의 행이나 열을 삭제하는 drop() 메소드를 제공.

drop() 메소드로 열을 삭제하려면 첫번째 매개변수에서 삭제하려는 열 이름을 전달하고 axis 매개변수를 1로 지정!

```python
ns_book = ns_df.drop('Unnamed: 13', axis = 1) # 삭제하려는 열 이름, axis 매개변수에 1을 지정하면 열을 삭제!
ns_book.head()
```

```python
ns_book = ns_df.drop(['부가기호', 'Unnamed: 13'], axis = 1)
ns_book.head()
```

지금까지는 ns_df 데이터프레임에서 열을 삭제한 결과를 새로운 ns_book이라는 변수에 저장했었는데,

drop() 메소드에 inplace 매개변수를 True로 지정하면 현재 선택한 데이터프레임을 바로 수정할 수 있다.

앞서 생성한 ns_book 데이터프레임의 '주제분류번호' 열을 삭제해보자.

```python
ns_book.drop('주제분류번호', axis = 1, inplace = True) # 선택한 데이터프레임에 덮어쓰기
ns_book.head()
```

새로운 변수를 만들어 데이터프레임 수정 결과를 반환하지 않고도 ns_book 데이터프레임에서 '주제분류번호' 열을 삭제!

판다스는 비어있는 값을 NaN으로 표시, drop() 메소드와 비슷한 dropna() 메소드는 기본적으로 NaN이 하나 이상 포함된 행이나 열을 삭제한다.

axis = 1로 지정하여 ns_df 데이터프레임에서 NaN이 포함된 열을 삭제!

14개의 열이 5개로 확 줄어들었다. 삭제된 열에는 NaN이 적어도 하나 이상 있었나 보다...

```python
ns_book = ns_df.dropna(axis = 1)
ns_book.head()
```

모든 값이 NaN인 열을 삭제하려면 dropna() 메소드에 how 매개변수를 'all'로 지정하면 된다.

```python
ns_book = ns_df.dropna(axis = 1, how = 'all')
ns_book.head()
```

지금까지 열을 삭제하는 방법을 알아보았다.

loc 메소드로 필요한 열을 선택하려면 슬라이스 연산자나 불리언 배열을 사용할 수 있다.

하지만 이 두가지가 유일한 방법은 아니다.

drop() 메소드로 열을 삭제하려면 axis = 1과 삭제하려는 열을 지정! 삭제할 열이 여러개일 때는 열 이름을 리스트로 전달할 수 있다.

또는 dropna() 메소드에 axis = 1을 지정하면 NaN이 들어있는 열을 삭제할 수 있다.

## 행 삭제하기

이번에는 행을 삭제하기. 행을 삭제할 때는 drop() 메소드를 사용할 수 있다.

axis 매개변수를 0으로 지정하면 행을 삭제할 수 있지만, 기본값이 0이기 때문에 생략해도 괜찮다.

처음에는 2개 행을 삭제하기!

```python
ns_book2 = ns_book.drop([0, 1])
ns_book2.head()
```

행을 선택할 때 자주 사용하는 [] 연산자를 살펴보기.

[] 연산자에 열 이름 또는 열 이름 리스트를 전달하여 데이터프레임의 열을 선택할 수 있었다.

```python
ns_book2 = ns_book[2:]
ns_book2.head()
```

슬라이싱 외에도 불리언 배열을 사용하여 행을 선택할 수 있다. 행을 선택할 때 가장 즐겨 사용하는 방법이다.

비교 연산자를 활용해 원하는 행은 True로 표시하고, 제외할 행은 False로 표시한 불리언 배열을 만들어 사용!

예를 들어 출판사가 '한빛미디어'인 행만 선택해 보자.

비교 연산자를 사용하여 원본 데이터프레임 ns_df에서 출판사가 '한빛미디어'인 행만 True인 불리언 배열을 만든다.

그 다음에 이 불리언 배열을 [] 연산자에 넣어 준다.

```python
selected_rows = ns_df['출판사'] == '한빛미디어'
ns_book2 = ns_df[selected_rows]
ns_book2.head()
```

loc 메소드에 불리언 배열을 사용하여 행을 선택할 수 있다!

```python
ns_book2 = ns_book.loc[selected_rows]
ns_book2.head()
```

한가지 예로, 대출건수가 1,000 이하인 행을 모두 삭제하고 싶다면 반대로

ns_book['대출건수'] > 1000 조건을 넣어 대출건수가 1,000이 넘는 행을 선택하면 된다!

```python
ns_book2 = ns_book[ns_book['대출건수'] > 1000]
ns_book2.head()
```

지금까지 불필요한 행을 삭제하거나 제외하는 방법을 알아보았다.

열을 삭제할 때와 마찬가지로 행을 삭제할 때도 loc 메소드, drop() 메소드, dropna() 메소드를 모두 사용할 수 있지만,

**불리언 배열을 [] 연산자에 전달하는 방법을 자주 사용한다는 것을 기억하자!!**

## 중복된 행 찾기

판다스 데이터프레임의 중복된 행은 duplicated() 메소드를 사용하여 검사할 수 있다.

중복된 행 중에서 처음 행을 제외한 나머지 행은 True로,

그 외에 중복되지 않은 나머지 모든 행은 False로 표시한 불리언 배열을 반환한다.

```python
sum(ns_book.duplicated())
```

sum() 함수가 반환한 값이 0인 것을 보니 True로 표시된 행이 하나도 없다! -> 중복된 행이 없다는 뜻!

이처럼 duplicated() 메소드는 기본적으로 데이터프레임에 있는 모든 열을 기준으로 중복된 행을 찾는다!

```python
sum(ns_book.duplicated(subset = ['도서명', '저자', 'ISBN']))
```

사실 ns_book 데이터프레임의 모든 행은 '번호' 열에 고유한 값을 가지고 있기 때문에 중복된 행이 나올 수 없다.

만약 '도서명', '저자', 'ISBN'을 기준으로 중복된 행이 있는지를 찾아보면?

일부 열을 기준으로 중복된 행을 찾으려면 duplicated() 메소드 안에 있는 **subset 매개변수**에 기준열을 나열한다!

'도서명', '저자', 'ISBN'을 기준으로 찾으면 중복된 행이 2만개가 넘는다!!

'도서명', '저자', 'ISBN'을 기준으로 어떤 데이터가 중복되었는지 확인해보자!

duplicated() 메소드에 keep 매개변수를 False로 지정하여 중복된 모든 행을 True로 표시한다.

이 데이터를 ns_book3 변수에 저장하여 확인해보자!

```python
dup_rows = ns_book.duplicated(subset = ['도서명', '저자', 'ISBN'], keep = False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

실행 결과를 확인해보니 첫번째 책 '파친코'는 두 권짜리 도서이다. 그래서 '권' 열에 1, 2로 표시된다.

두번째 '보건교사 안은영'은 한 권짜리 도서인데, 두 권이 등록되어 있다. 아마도 인기가 많아 도서관에서 두권을 구입한 것으로 보인다.

## 그룹별로 모으기

어떤 도서가 인기 있을지 예상하려고 하므로 이 데이터 프레임에서 '대출건수'열이 중요!!


같은 도서의 대출건수를 하나로 합치기!!
예를 들어 '혼공 데이터 분석' 도서가 두 권이라면 하나로 합치려고 한다.


이럴때 groupby() 메소드를 사용! groupby() 메소드의 'by' 매개변수에는 행을 합칠 때 기준이 되는 열을 지정!


권수가 한개 이상인 도서는 각기 다른 도서로 간주한다. 예를 들어 파친코 1권과 2권을 별개의 도서로 본다.


'도서명', '출판사', '발행연도', '대출건수' 열을 기준으로 행을 합치면 다음과 같다.

```python
count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']] # 그룹으로 묶을 기준 열
print(count_df)
```

결과를 변환한 count_df 데이터프레임에 groupby( ) 메소드를 적용.


정수 타입인 '대출건수' 열은 sum( ) 메소드를 사용,

NaN이 포함되어 있는 행을 삭제하고 계산하면 대출건수 합계에서 빠지기 때문에 이를 막기 위해 dropna 매개변수를 False로 지정

```python
group_df = count_df.groupby(by = ['도서명', '저자', 'ISBN', '권'], dropna = False)
loan_count = group_df.sum()
print(loan_count)
```

```python
loan_count = count_df.groupby(by = ['도서명', '저자', 'ISBN', '권'], dropna = False).sum()
loan_count.head()
```
